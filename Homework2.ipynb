##### Question 2 (10 points)

Recreate the figure in Part 3 Question 3 in Homework 1 with Altair. Include necessary title and labels. Here is the question again:
    
    Generate a data visualization to show the relationship between max temperature and min temperature, and then show different color for differnet weather type.

```python
import altair as alt
from vega_datasets import data

weather = data.seattle_weather()

# Define custom colors for different weather types
weather_colors = {
    'sun': 'orange',
    'rain': 'blue',
    'snow': 'cyan',
    'drizzle': 'green',
    'fog': 'gray'
}

# Plot each weather type with its respective color
scatter = alt.Chart(weather).mark_circle(size=50, opacity=0.7).encode(
    x=alt.X('temp_min:Q', title='Min Temperature (°C)'),  
    y=alt.Y('temp_max:Q', title='Max Temperature (°C)'), 
    color=alt.Color('weather:N', 
                   title='Weather Type', 
                   scale=alt.Scale(
                       domain=list(weather_colors.keys()),  
                       range=list(weather_colors.values())  
                   )),
    tooltip=['weather:N', 'temp_min:Q', 'temp_max:Q']  
).properties(
    title='Max vs Min Temperature by Weather Type',
    width=600,
    height=400
).configure_axis(
    grid=True  
)

scatter
```

##### Question 3 (10 points)

Create a facet figure in Altair with 2 row and 3 columns. For each panel in the facet figure, it shows a scatter plot between the max and min temperature for one type of weather. Use the wind speed to color the points. Also add interaction to show the max and min temperature and wind speed for each data point. Include necessary title and labels. 

```python
import altair as alt
from vega_datasets import data

weather = data.seattle_weather()

base = alt.Chart(weather).mark_circle(size=50, opacity=0.7).encode(
    x=alt.X('temp_min:Q', title='Min Temperature (°C)'),
    y=alt.Y('temp_max:Q', title='Max Temperature (°C)'),
    color=alt.Color('wind:Q', title='Wind Speed', scale=alt.Scale(scheme='viridis')),
    tooltip=[
        alt.Tooltip('temp_min:Q', title='Min Temp (°C)'),
        alt.Tooltip('temp_max:Q', title='Max Temp (°C)'),
        alt.Tooltip('wind:Q', title='Wind Speed', format='.2f')
    ]
).properties(
    width=200,
    height=150
)

# Create a facet figure in Altair with 2 row and 3 columns. 
facet_chart = base.facet(
    facet=alt.Column(
        'weather:N',
        header=alt.Header(
            title='Weather Type',   
            titleOrient='bottom',    
            labelOrient='bottom'     
        )
    ),
    columns=3   
).properties(
    title='Max vs Min Temperature by Weather Type'
).resolve_scale(
    color='shared'
).configure_header(
    labelFontSize=12,
    titleFontSize=14
).configure_view(
    stroke='transparent'
)

facet_chart

```

The type of illusion I designed is: Simultaneous Contrast.

Design Concept: This illusion demonstrates how the same color appears different depending on the background. I placed identical gray squares on a dark background and a light background. The human eye perceives the gray on the dark background as brighter and the gray on the light background as darker.

Cause of the Illusion: Due to the contrast adaptation mechanism of human vision, we tend to overestimate brightness when a color is placed on a dark background and underestimate brightness when it is placed on a light background.
